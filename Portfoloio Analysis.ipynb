```python
from datetime import datetime, timezone
import requests

#--------------------- FUNCTIONS ---------------------------------------------#

#----------- Prior to stock details----------#


def fetch_stock_details(ticker):
    url = "https://yahoo-finance15.p.rapidapi.com/api/v1/markets/stock/history"
    querystring = {"symbol":ticker,"interval":"3mo","diffandsplits":"false"}
    headers = {
        "x-rapidapi-key": "REMOVED",
        "x-rapidapi-host": "yahoo-finance15.p.rapidapi.com"
    }
    
    response = requests.get(url, headers=headers, params=querystring)
    stock_json = response.json()
    
    return stock_json

def organize_and_extract(stock_json):
    stock_dataframe = pd.DataFrame(stock_json["body"]).T.reset_index(drop=True) # convert, transpose and drop index
    
    return stock_dataframe
    
def check_if_in_file(ticker):
    # Load the Excel file
    file_name = "stock_details.xlsx"
    excel_file = pd.ExcelFile(file_name)
    
    # Get the list of sheet names
    sheet_names = excel_file.sheet_names
    if ticker in sheet_names:
        return True
    else:
        return False

# ------------- after gotten the stock details---------------#

def save_stock_details(ticker, dataframe):
    with pd.ExcelWriter("stock_details.xlsx", mode='a', if_sheet_exists='replace') as writer:
        dataframe.to_excel(writer, sheet_name=ticker, index=False)
    
    print(f"{ticker} details saved!")

def date_to_epoch(date):
    # Parse the date string and set timezone to UTC
    dt = datetime.strptime(date_str, "%d-%m-%Y").replace(tzinfo=timezone.utc)
    
    # Convert to Unix Epoch
    epoch_time = int(dt.timestamp())
    
    return epoch_time


def get_price(date):
    date_epoch = date_to_epoch(date)
    # Find the index of the closest point to x_target
    index = np.abs(x_ - date_epoch).argmin()
    y_interpolated = y_[index]
    
    return y_interpolated
    
```
