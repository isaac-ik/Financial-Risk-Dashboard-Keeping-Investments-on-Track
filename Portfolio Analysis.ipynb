```python
from datetime import datetime, timezone
import requests

#--------------------- FUNCTIONS ---------------------------------------------#

#----------- Prior to stock details----------#


def fetch_stock_details(ticker):
    url = "https://yahoo-finance15.p.rapidapi.com/api/v1/markets/stock/history"
    querystring = {"symbol":ticker,"interval":"3mo","diffandsplits":"false"}
    headers = {
        "x-rapidapi-key": "REMOVED",
        "x-rapidapi-host": "yahoo-finance15.p.rapidapi.com"
    }
    
    response = requests.get(url, headers=headers, params=querystring)
    stock_json = response.json()
    
    return stock_json

def organize_and_extract(stock_json):
    stock_dataframe = pd.DataFrame(stock_json["body"]).T.reset_index(drop=True) # convert, transpose and drop index
    
    return stock_dataframe
    
def check_if_in_file(ticker):
    # Load the Excel file
    file_name = "stock_details.xlsx"
    excel_file = pd.ExcelFile(file_name)
    
    # Get the list of sheet names
    sheet_names = excel_file.sheet_names
    if ticker in sheet_names:
        return True
    else:
        return False

# ------------- after gotten the stock details---------------#

def save_stock_details(ticker, dataframe):
    with pd.ExcelWriter("stock_details.xlsx", mode='a', if_sheet_exists='replace') as writer:
        dataframe.to_excel(writer, sheet_name=ticker, index=False)
    
    print(f"{ticker} details saved!")

def date_to_epoch(date):
    # Parse the date string and set timezone to UTC
    dt = datetime.strptime(date_str, "%d-%m-%Y").replace(tzinfo=timezone.utc)
    
    # Convert to Unix Epoch
    epoch_time = int(dt.timestamp())
    
    return epoch_time


def get_price(date):
    date_epoch = date_to_epoch(date)
    # Find the index of the closest point to x_target
    index = np.abs(x_ - date_epoch).argmin()
    y_interpolated = y_[index]
    
    return y_interpolated
    
```

```python
import pandas as pd

# Read stock names from the first Excel file
stock_file = "list_of_assets.xlsx"
stock_df = pd.read_excel(stock_file)

# Extract the list of stock names
stock_names = stock_df['asset_name'].tolist()

# Initialize the portfolio dictionary
portfolio = {}

# Read stock details for each stock from the second Excel file
details_file = "stock_details.xlsx"

for stockname in stock_names:
    # Load the sheet with the stock name
    if stockname == 'FB':
        continue
        
    stock_data = pd.read_excel(details_file, sheet_name=stockname)
    
    # Initialize sub-dictionaries for each stock
    portfolio[stockname] = {
        "date": {},
        "date_utc": {},
        "close": {}
    }
    
    # Populate the portfolio with the extracted data
    for idx, row in stock_data.iterrows():
        date_str = row['date']
        date_utc = row['date_utc']
        close_price = row['close']
        
        # Use the index as the unique key in each dictionary
        portfolio[stockname]["date"][idx] = date_str
        portfolio[stockname]["date_utc"][idx] = date_utc
        portfolio[stockname]["close"][idx] = close_price

# Print the portfolio dictionary (for verification)
print(portfolio)
```

```python
from scipy.interpolate import interp1d
import numpy as np

def interpolate_stock_details(portfolio):
    interpolated_data = {}

    for stockname, details in portfolio.items():
        # Extract x (date_utc) and y (close) values from the portfolio
        x = np.array(list(details["date_utc"].values()))
        y = np.array(list(details["close"].values()))
        
        # Ensure enough data points for cubic interpolation
        if len(x) < 4:  # Cubic interpolation requires at least 4 points
            print(f"Not enough data points for stock: {stockname}")
            continue
        
        # Define the index positions
        n = np.arange(len(x))

        # Perform cubic interpolation for both x and y
        x_spline = interp1d(n, x, kind='cubic')
        y_spline = interp1d(n, y, kind='cubic')

        # Interpolation over a denser range of n values
        n_ = np.linspace(n.min(), n.max(), 2545)
        x_ = x_spline(n_)
        y_ = y_spline(n_)

        # Store the interpolated results in the dictionary
        interpolated_data[stockname] = {"x_": x_.tolist(), "y_": y_.tolist()}
    
    return interpolated_data


```

```python
import matplotlib.pyplot as plt

def plot_stock_comparison(portfolio, interpolated_data):
    for stockname in interpolated_data:
        # Extract original data
        x_original = list(portfolio[stockname]["date_utc"].values())
        y_original = list(portfolio[stockname]["close"].values())

        # Extract interpolated data
        x_interpolated = interpolated_data[stockname]["x_"]
        y_interpolated = interpolated_data[stockname]["y_"]

        # Plot original data and interpolated spline
        plt.figure(figsize=(10, 6))
        plt.plot(x_interpolated, y_interpolated, label="Interpolated Spline", color="orange")
        plt.plot(x_original, y_original, color="blue", label="Original Data Points")
        
        # Title and labels
        plt.title(f"{stockname} - Original vs Interpolated Data")
        plt.xlabel("Date (UTC)")
        plt.ylabel("Close Price")
        plt.legend()
        plt.grid(True)
        plt.show()

```
